````python
import warnings
import torch

# MONAI imports
from monai.transforms import (
    Compose, LoadImage, ScaleIntensityRanged, CropForegroundd,
    Spacingd, EnsureTyped, EnsureChannelFirstd
)
from monai.inferers import sliding_window_inference
from monai.networks.nets import SwinUNETR
from monai.utils import set_determinism

# TotalSegmentator
try:
    from totalsegmentator.python_api import totalsegmentator
    TOTALSEGMENTATOR_AVAILABLE = True
except ImportError:
    TOTALSEGMENTATOR_AVAILABLE = False
    print(" TotalSegmentator not available")

warnings.filterwarnings("ignore")
set_determinism(seed=42)

# Check GPU availability
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f" Using device: {device}")
if torch.cuda.is_available():
    print(f"   GPU: {torch.cuda.get_device_name(0)}")
    print(f"   Memory: {torch.cuda.get_device_properties(0).total_memory / 1024**3:.2f} GB")
````

````python
from enum import Enum
from dataclasses import dataclass
from typing import Tuple

# ============================================================================
# CELL 3: Configuration Classes
# ============================================================================

class ModelType(Enum):
    """Model types"""
    SWIN_UNETR = "swin_unetr"
    TOTALSEGMENTATOR = "totalsegmentator"

@dataclass
class ColabConfig:
    """Simplified configuration for Colab"""
    # Paths (modify these to your paths)
    output_dir: str = "/content/output"
    swin_model_path: str = None  # Set to your model path

    # Processing settings (optimized for Colab)
    roi_size: Tuple[int, int, int] = (96, 96, 96)
    sw_batch_size: int = 2
    overlap: float = 0.5

    # Models to use
    use_swin: bool = True
    use_totalsegmentator: bool = True

    # Device
    device: str = "cuda" if torch.cuda.is_available() else "cpu"

# Create default config
config = ColabConfig()
print(f" Configuration created")
print(f"   Output: {config.output_dir}")
print(f"   ROI size: {config.roi_size}")
````

````python
# ============================================================================
# CELL 4: Organ Mapping
# ============================================================================

class OrganMapping:
    """Organ label mappings"""

    BTCV_ORGANS = {
        1: "spleen",
        2: "kidney_right",
        3: "kidney_left",
        4: "gallbladder",
        5: "esophagus",
        6: "liver",
        7: "stomach",
        8: "aorta",
        9: "inferior_vena_cava",
        10: "portal_vein_and_splenic_vein",
        11: "pancreas",
        12: "adrenal_gland_right",
        13: "adrenal_gland_left",
    }

    TS_TO_BTCV = {
        'spleen.nii.gz': 1,
        'kidney_right.nii.gz': 2,
        'kidney_left.nii.gz': 3,
        'gallbladder.nii.gz': 4,
        'esophagus.nii.gz': 5,
        'liver.nii.gz': 6,
        'stomach.nii.gz': 7,
        'aorta.nii.gz': 8,
        'inferior_vena_cava.nii.gz': 9,
        'portal_vein_and_splenic_vein.nii.gz': 10,
        'pancreas.nii.gz': 11,
        'adrenal_gland_right.nii.gz': 12,
        'adrenal_gland_left.nii.gz': 13,
    }

organ_map = OrganMapping()
print(f" Organ mapping loaded: {len(organ_map.BTCV_ORGANS)} organs")
````

````python
from typing import Dict
from pathlib import Path
import gc
import nibabel as nib

# ============================================================================
# CELL 5: Utility Functions
# ============================================================================

def setup_output_dirs(base_dir: str) -> Dict[str, Path]:
    """Create output directory structure"""
    base = Path(base_dir)
    dirs = {
        'swin': base / 'swin_unetr',
        'swin_organs': base / 'swin_unetr' / 'organs',
        'ts': base / 'totalsegmentator',
        'ts_organs': base / 'totalsegmentator' / 'organs',
        'ensemble': base / 'ensemble',
        'visualizations': base / 'visualizations'
    }

    for dir_path in dirs.values():
        dir_path.mkdir(parents=True, exist_ok=True)

    return dirs

def cleanup_memory():
    """Clean up GPU memory"""
    gc.collect()
    if torch.cuda.is_available():
        torch.cuda.empty_cache()

def validate_image(image_path: str) -> bool:
    """Validate NIfTI image"""
    try:
        if not Path(image_path).exists():
            print(f" File not found: {image_path}")
            return False

        img = nib.load(image_path)
        if len(img.shape) < 3:
            print(f" Invalid dimensions: {img.shape}")
            return False

        print(f" Valid image: {img.shape}, {img.get_fdata().dtype}")
        return True
    except Exception as e:
        print(f" Validation error: {e}")
        return False

# Setup output directories
output_dirs = setup_output_dirs(config.output_dir)
print(f" Output directories created")
````

````python
# ============================================================================
# CELL 6: Preprocessing
# ============================================================================

def create_preprocessing_transforms():
    """Create MONAI preprocessing transforms"""
    transforms = Compose([
        EnsureChannelFirstd(keys=["image", "label"]),
        ScaleIntensityRanged(
            keys=["image"],
            a_min=-175,
            a_max=250,
            b_min=0.0,
            b_max=1.0,
            clip=True
        ),
        Spacingd(
            keys=["image", "label"],
            pixdim=(1.5, 1.5, 2.0),
            mode=("bilinear", "nearest"),
        ),
        CropForegroundd(
            keys=["image", "label"],
            source_key="image",
            allow_smaller=True
        ),
        EnsureTyped(
            keys=["image", "label"],
            device=device,
            track_meta=True
        )
    ])
    return transforms

def preprocess_image(image_path: str, transforms):
    """Preprocess a single image"""
    print(f" Preprocessing {Path(image_path).name}...")

    # Load original for reference
    original_nifti = nib.load(image_path)
    original_shape = original_nifti.shape
    original_affine = original_nifti.affine

    # Load with MONAI
    loader = LoadImage(image_only=True, ensure_channel_first=True)
    image = loader(image_path)

    # Handle dimensions
    if image.dim() == 4 and image.shape[0] == 1:
        image = image.squeeze(0)

    # Create data dict
    data = {"image": image, "label": image}

    # Apply transforms
    data = transforms(data)

    print(f"   Original shape: {original_shape}")
    print(f"   Processed shape: {data['image'].shape}")

    return {
        'data': data,
        'original_shape': original_shape,
        'original_affine': original_affine,
        'path': image_path
    }

# Create transforms
transforms = create_preprocessing_transforms()
print(" Preprocessing transforms created")
````

````python
from typing import Optional, Tuple
import time
import numpy as np

# ============================================================================
# CELL 7: Swin UNETR Model
# ============================================================================

def load_swin_unetr(model_path: Optional[str] = None):
    """Load Swin UNETR model"""
    print(" Loading Swin UNETR model...")

    model = SwinUNETR(
        in_channels=1,
        out_channels=14,
        feature_size=48,
        use_checkpoint=True,
    ).to(device)

    if model_path and Path(model_path).exists():
        print(f"   Loading weights from: {model_path}")
        weights = torch.load(model_path, map_location=device, weights_only=True)
        model.load_state_dict(weights)
        print("    Weights loaded")
    else:
        print("    No weights provided, using random initialization")

    model.eval()
    return model

def segment_with_swin(model, preprocessed_data, config):
    """Perform segmentation with Swin UNETR"""
    print(" Running Swin UNETR inference...")
    start_time = time.time()

    with torch.no_grad():
        input_tensor = preprocessed_data['data']['image'].unsqueeze(0).to(device)

        output = sliding_window_inference(
            input_tensor,
            roi_size=config.roi_size,
            sw_batch_size=config.sw_batch_size,
            predictor=model,
            overlap=config.overlap,
            mode="gaussian"
        )

        seg_volume = torch.argmax(output, dim=1).cpu().numpy()[0]

    elapsed = time.time() - start_time
    print(f"    Completed in {elapsed:.2f}s")
    print(f"   Found organs: {np.unique(seg_volume)}")

    return seg_volume, elapsed

def resize_segmentation(seg_volume: np.ndarray, target_shape: Tuple) -> np.ndarray:
    """Resize segmentation to original shape"""
    from scipy import ndimage

    if seg_volume.shape == target_shape:
        return seg_volume

    zoom_factors = [t / s for t, s in zip(target_shape, seg_volume.shape)]
    resized = ndimage.zoom(seg_volume, zoom_factors, order=0, mode='nearest')

    return resized.astype(np.uint8)

# Load model (if you have weights)
swin_model = None
if config.use_swin:
    swin_model = load_swin_unetr(config.swin_model_path)
    print(" Swin UNETR ready")
````

````python
# ============================================================================
# CELL 8: TotalSegmentator
# ============================================================================

def segment_with_totalsegmentator(image_path: str, output_dir: Path):
    """Perform segmentation with TotalSegmentator"""
    if not TOTALSEGMENTATOR_AVAILABLE:
        print(" TotalSegmentator not available")
        return None, 0

    print(" Running TotalSegmentator...")
    start_time = time.time()

    # Create temp directory
    import tempfile
    temp_dir = Path(tempfile.mkdtemp())

    try:
        # Run TotalSegmentator
        totalsegmentator(str(image_path), str(temp_dir), fast=False)

        # Load original for reference
        original_nifti = nib.load(image_path)
        original_shape = original_nifti.shape

        # Find reference file
        reference_file = next(temp_dir.glob('*.nii.gz'), None)
        if not reference_file:
            print(" No output files found")
            return None, 0

        # Load reference
        ref_img = nib.load(reference_file)
        combined_seg = np.zeros(ref_img.shape, dtype=np.uint8)

        # Combine organs
        organs_found = []
        for ts_filename, label_id in organ_map.TS_TO_BTCV.items():
            organ_path = temp_dir / ts_filename
            if organ_path.exists():
                organ_data = nib.load(organ_path).get_fdata() > 0
                combined_seg[organ_data] = label_id
                organs_found.append(organ_map.BTCV_ORGANS[label_id])

        elapsed = time.time() - start_time
        print(f"    Completed in {elapsed:.2f}s")
        print(f"   Found {len(organs_found)} organs: {', '.join(organs_found[:3])}...")

        return combined_seg, elapsed

    finally:
        # Cleanup
        import shutil
        shutil.rmtree(temp_dir, ignore_errors=True)
````

````python
# ============================================================================
# CELL 9: Save Functions
# ============================================================================

def save_segmentation(seg_volume: np.ndarray, affine: np.ndarray,
                     patient_id: str, model_name: str, output_dirs: Dict):
    """Save segmentation outputs"""
    print(f" Saving {model_name} results for {patient_id}...")

    # Save combined segmentation
    output_path = output_dirs[model_name] / f"{patient_id}_segmentation.nii.gz"
    nifti_img = nib.Nifti1Image(seg_volume.astype(np.uint8), affine)
    nib.save(nifti_img, str(output_path))
    print(f"   Saved: {output_path}")

    # Save individual organs
    organs_dir = output_dirs[f'{model_name}_organs']
    organs_saved = []

    for organ_id, organ_name in organ_map.BTCV_ORGANS.items():
        organ_mask = (seg_volume == organ_id).astype(np.uint8)
        if organ_mask.any():
            organ_path = organs_dir / f"{patient_id}_{organ_name}.nii.gz"
            organ_nifti = nib.Nifti1Image(organ_mask, affine)
            nib.save(organ_nifti, str(organ_path))
            organs_saved.append(organ_name)

    print(f"   Saved {len(organs_saved)} individual organs")
    return True
````

````python
import matplotlib.pyplot as plt
from typing import Optional
import numpy as np
from pathlib import Path

# ============================================================================
# CELL 10: Visualization
# ============================================================================

def visualize_segmentation(image_path: str, segmentations: Dict,
                          slice_idx: Optional[int] = None,
                          save_path: Optional[str] = None):
    """Visualize segmentation results"""
    print(" Creating visualization...")

    # Load original image
    loader = LoadImage(image_only=True)
    image = loader(image_path)
    if image.ndim == 4:
        image = image[0]
    image = image.numpy()

    # Determine slice
    if slice_idx is None:
        slice_idx = image.shape[2] // 2

    # Create figure
    num_models = len(segmentations) + 1
    fig, axes = plt.subplots(1, num_models, figsize=(5 * num_models, 5))

    if num_models == 2:
        axes = [axes] if not isinstance(axes, np.ndarray) else axes

    # Plot original
    axes[0].imshow(image[:, :, slice_idx], cmap='gray')
    axes[0].set_title('Original Image')
    axes[0].axis('off')

    # Plot segmentations
    for idx, (model_name, seg) in enumerate(segmentations.items(), 1):
        axes[idx].imshow(seg[:, :, slice_idx], cmap='nipy_spectral', vmin=0, vmax=13)
        axes[idx].set_title(f'{model_name.upper()}')
        axes[idx].axis('off')

    patient_id = Path(image_path).stem.replace('.nii', '')
    plt.suptitle(f'Patient: {patient_id} - Slice: {slice_idx}/{image.shape[2]}', fontsize=16)
    plt.tight_layout()

    if save_path:
        plt.savefig(save_path, dpi=150, bbox_inches='tight')
        print(f"   Saved: {save_path}")
    else:
        plt.show()

    plt.close()

def create_comparison_grid(image_path: str, segmentations: Dict,
                          num_slices: int = 5, save_path: Optional[str] = None):
    """Create multi-slice comparison grid"""
    print(" Creating comparison grid...")

    # Load image
    loader = LoadImage(image_only=True)
    image = loader(image_path)
    if image.ndim == 4:
        image = image[0]
    image = image.numpy()

    # Select slices
    depth = image.shape[2]
    slice_indices = np.linspace(depth // 4, 3 * depth // 4, num_slices, dtype=int)

    # Create grid
    num_models = len(segmentations) + 1
    fig, axes = plt.subplots(num_slices, num_models, figsize=(5 * num_models, 4 * num_slices))

    for row, slice_idx in enumerate(slice_indices):
        # Original image
        axes[row, 0].imshow(image[:, :, slice_idx], cmap='gray')
        axes[row, 0].set_title(f'Original - Slice {slice_idx}')
        axes[row, 0].axis('off')

        # Segmentations
        for col, (model_name, seg) in enumerate(segmentations.items(), 1):
            axes[row, col].imshow(seg[:, :, slice_idx], cmap='nipy_spectral', vmin=0, vmax=13)
            if row == 0:
                axes[row, col].set_title(f'{model_name.upper()}')
            axes[row, col].axis('off')

    patient_id = Path(image_path).stem.replace('.nii', '')
    plt.suptitle(f'Patient: {patient_id} - Multi-Slice Comparison', fontsize=16)
    plt.tight_layout()

    if save_path:
        plt.savefig(save_path, dpi=150, bbox_inches='tight')
        print(f"   Saved: {save_path}")
    else:
        plt.show()

    plt.close()
````

````python
# ============================================================================
# CELL 11: Main Processing Function
# ============================================================================

def process_patient(image_path: str, config: ColabConfig,
                   models: Dict, output_dirs: Dict) -> Dict:
    """Process a single patient with all models"""
    patient_id = Path(image_path).stem.replace('.nii', '')

    print("\n" + "="*70)
    print(f" Processing Patient: {patient_id}")
    print("="*70)

    # Validate
    if not validate_image(image_path):
        return {'error': 'Invalid image'}

    results = {
        'patient_id': patient_id,
        'image_path': image_path,
        'segmentations': {},
        'times': {}
    }

    # Load original
    original_nifti = nib.load(image_path)
    original_affine = original_nifti.affine
    original_shape = original_nifti.shape

    # Process with Swin UNETR
    if config.use_swin and 'swin' in models:
        try:
            preprocessed = preprocess_image(image_path, transforms)
            seg_volume, elapsed = segment_with_swin(models['swin'], preprocessed, config)

            # Resize to original
            seg_volume = resize_segmentation(seg_volume, original_shape)

            # Save
            save_segmentation(seg_volume, original_affine, patient_id, 'swin', output_dirs)

            results['segmentations']['swin'] = seg_volume
            results['times']['swin'] = elapsed

            cleanup_memory()

        except Exception as e:
            print(f" Swin UNETR failed: {e}")

    # Process with TotalSegmentator
    if config.use_totalsegmentator and TOTALSEGMENTATOR_AVAILABLE:
        try:
            seg_volume, elapsed = segment_with_totalsegmentator(image_path, output_dirs['ts'])

            if seg_volume is not None:
                # Save
                save_segmentation(seg_volume, original_affine, patient_id, 'ts', output_dirs)

                results['segmentations']['ts'] = seg_volume
                results['times']['ts'] = elapsed

            cleanup_memory()

        except Exception as e:
            print(f" TotalSegmentator failed: {e}")

    # Create visualizations
    if results['segmentations']:
        vis_path = output_dirs['visualizations'] / f"{patient_id}_comparison.png"
        visualize_segmentation(image_path, results['segmentations'], save_path=str(vis_path))

        grid_path = output_dirs['visualizations'] / f"{patient_id}_grid.png"
        create_comparison_grid(image_path, results['segmentations'], save_path=str(grid_path))

    # Summary
    print("\n Processing Summary:")
    for model_name, elapsed in results['times'].items():
        print(f"   {model_name.upper()}: {elapsed:.2f}s")

    return results
````

````python
from typing import List, Dict
# ============================================================================
# CELL 12: Batch Processing
# ============================================================================

def process_batch(image_paths: List[str], config: ColabConfig,
                 models: Dict, output_dirs: Dict) -> List[Dict]:
    """Process multiple patients"""
    print(f"\n Starting batch processing: {len(image_paths)} patients")
    print("="*70)

    results = []
    start_time = time.time()

    for idx, image_path in enumerate(image_paths, 1):
        print(f"\n[{idx}/{len(image_paths)}] {Path(image_path).name}")

        try:
            result = process_patient(image_path, config, models, output_dirs)
            results.append(result)
        except Exception as e:
            print(f" Failed: {e}")
            results.append({'error': str(e), 'image_path': image_path})

        # Memory cleanup every 3 patients
        if idx % 3 == 0:
            cleanup_memory()

    # Final summary
    total_time = time.time() - start_time
    successful = sum(1 for r in results if 'error' not in r)

    print("\n" + "="*70)
    print(" Batch Processing Complete!")
    print(f"   Total time: {total_time:.2f}s")
    print(f"   Average: {total_time/len(image_paths):.2f}s per patient")
    print(f"   Success: {successful}/{len(image_paths)}")
    print("="*70)

    return results
````

````python
# ============================================================================
# CELL: WholeBodyCT Integration
# ============================================================================

print(" Adding WholeBodyCT support...")

# Import additional dependencies
from monai.bundle import ConfigParser

# ============================================================================
# Update Configuration
# ============================================================================

# Add WholeBodyCT paths to config
config.wbct_model_path = "/content/wholebody_ct_model/wholeBody_ct_segmentation/models/model.pt"
config.wbct_config_path = "/content/wholebody_ct_model/wholeBody_ct_segmentation/configs/inference.json"
config.use_wholebody_ct = True

print(" Configuration updated with WholeBodyCT options")

# ============================================================================
# Update Organ Mapping
# ============================================================================

# Add WholeBodyCT to BTCV mapping
organ_map.WBCT_TO_BTCV_INDICES = {
    1: 1,   # spleen -> spleen
    2: 2,   # right_kidney -> kidney_right
    3: 3,   # left_kidney -> kidney_left
    4: 4,   # gallbladder -> gallbladder
    5: 5,   # esophagus -> esophagus
    6: 6,   # liver -> liver
    7: 7,   # stomach -> stomach
    8: 8,   # aorta -> aorta
    9: 9,   # inferior_vena_cava -> inferior_vena_cava
    10: 10, # portal_vein_and_splenic_vein -> portal_vein_and_splenic_vein
    11: 11, # pancreas -> pancreas
    12: 12, # right_adrenal_gland -> adrenal_gland_right
    13: 13, # left_adrenal_gland -> adrenal_gland_left
}

print(" Organ mapping updated")

# ============================================================================
# Update Output Directories
# ============================================================================

# Add WholeBodyCT output directories
output_dirs['wbct'] = Path(config.output_dir) / 'wholebody_ct'
output_dirs['wbct_organs'] = Path(config.output_dir) / 'wholebody_ct' / 'organs'

# Create directories
output_dirs['wbct'].mkdir(parents=True, exist_ok=True)
output_dirs['wbct_organs'].mkdir(parents=True, exist_ok=True)

print(" Output directories created for WholeBodyCT")

# ============================================================================
# WholeBodyCT Functions
# ============================================================================

def load_wholebody_ct(model_path: Optional[str] = None,
                      config_path: Optional[str] = None):
    """Load WholeBodyCT model from MONAI bundle"""
    if not config_path or not Path(config_path).exists():
        print(" WholeBodyCT config not found")
        return None, None

    print(" Loading WholeBodyCT model...")

    try:
        # Load configuration
        config_parser = ConfigParser()
        config_parser.read_config(config_path)

        # Get components
        preprocessing = config_parser.get_parsed_content("preprocessing")
        model = config_parser.get_parsed_content("network")
        inferer = config_parser.get_parsed_content("inferer")
        postprocessing = config_parser.get_parsed_content("postprocessing")

        # Load weights
        if model_path and Path(model_path).exists():
            print(f"   Loading weights from: {model_path}")
            weights = torch.load(model_path, map_location=device, weights_only=True)
            model.load_state_dict(weights)
            model.to(device)
            model.eval()
            print("    Weights loaded")
        else:
            print("    Model weights not found")
            return None, None

        # Store components
        model_components = {
            'model': model,
            'preprocessing': preprocessing,
            'inferer': inferer,
            'postprocessing': postprocessing,
            'config_parser': config_parser
        }

        return model, model_components

    except Exception as e:
        print(f"    Failed to load WholeBodyCT: {e}")
        import traceback
        traceback.print_exc()
        return None, None

def convert_wbct_to_btcv(wbct_output: np.ndarray) -> np.ndarray:
    """Convert WholeBodyCT output to BTCV format"""

    # Handle multi-channel output (probability maps)
    if wbct_output.ndim == 4:
        num_classes = wbct_output.shape[0]
        combined = np.zeros(wbct_output.shape[1:], dtype=np.uint8)

        # Map WholeBodyCT organs to BTCV indices
        for wbct_idx, btcv_idx in organ_map.WBCT_TO_BTCV_INDICES.items():
            if wbct_idx < num_classes:
                organ_mask = wbct_output[wbct_idx] > 0.5
                combined[organ_mask] = btcv_idx

        return combined

    # Handle single-channel output (class indices)
    elif wbct_output.ndim == 3:
        unique_values = np.unique(wbct_output)

        if len(unique_values) <= 20:  # Likely already class indices
            # Convert to BTCV format
            combined = np.zeros_like(wbct_output, dtype=np.uint8)
            for wbct_idx, btcv_idx in organ_map.WBCT_TO_BTCV_INDICES.items():
                mask = wbct_output == wbct_idx
                combined[mask] = btcv_idx
            return combined
        else:
            # Continuous values, use thresholding
            binary = (wbct_output > 0.5).astype(np.uint8)
            return binary

    return np.zeros_like(wbct_output, dtype=np.uint8)

def segment_with_wholebody_ct(image_path: str, model_components: Dict):
    """Perform segmentation with WholeBodyCT"""
    if model_components is None:
        print(" WholeBodyCT model not loaded")
        return None, 0

    print(" Running WholeBodyCT inference...")
    start_time = time.time()

    try:
        with torch.no_grad():
            # Preprocess
            data = model_components['preprocessing']({'image': image_path})

            # Inference
            data['pred'] = model_components['inferer'](
                data['image'].unsqueeze(0).to(device),
                network=model_components['model']
            )

            # Remove batch dimension
            data['pred'] = data['pred'][0]

            # Postprocessing
            data = model_components['postprocessing'](data)

            # Get segmentation
            if isinstance(data['pred'], torch.Tensor):
                segmentation = data['pred'].cpu().numpy()
            else:
                segmentation = data['pred']

            # Convert to BTCV format
            seg_volume = convert_wbct_to_btcv(segmentation)

            elapsed = time.time() - start_time
            print(f"    Completed in {elapsed:.2f}s")
            print(f"   Found organs: {np.unique(seg_volume)}")

            return seg_volume, elapsed

    except Exception as e:
        print(f"    WholeBodyCT failed: {e}")
        import traceback
        traceback.print_exc()
        return None, 0

print(" WholeBodyCT functions defined")

# ============================================================================
# Load WholeBodyCT Model
# ============================================================================

wbct_model = None
wbct_components = None

if config.use_wholebody_ct:
    wbct_model, wbct_components = load_wholebody_ct(
        config.wbct_model_path,
        config.wbct_config_path
    )
    if wbct_model is not None:
        print(" WholeBodyCT ready")
    else:
        print(" WholeBodyCT not available (set model_path and config_path)")

# ============================================================================
# Update Process Patient Function
# ============================================================================

def process_patient_with_wbct(image_path: str, config: ColabConfig,
                              models: Dict, output_dirs: Dict) -> Dict:
    """Enhanced process_patient function with WholeBodyCT support"""
    patient_id = Path(image_path).stem.replace('.nii', '')

    print("\n" + "="*70)
    print(f" Processing Patient: {patient_id}")
    print("="*70)

    # Validate
    if not validate_image(image_path):
        return {'error': 'Invalid image'}

    results = {
        'patient_id': patient_id,
        'image_path': image_path,
        'segmentations': {},
        'times': {}
    }

    # Load original
    original_nifti = nib.load(image_path)
    original_affine = original_nifti.affine
    original_shape = original_nifti.shape

    # Process with Swin UNETR
    if config.use_swin and 'swin' in models:
        try:
            preprocessed = preprocess_image(image_path, transforms)
            seg_volume, elapsed = segment_with_swin(models['swin'], preprocessed, config)

            # Resize to original
            seg_volume = resize_segmentation(seg_volume, original_shape)

            # Save
            save_segmentation(seg_volume, original_affine, patient_id, 'swin', output_dirs)

            results['segmentations']['swin'] = seg_volume
            results['times']['swin'] = elapsed

            cleanup_memory()

        except Exception as e:
            print(f" Swin UNETR failed: {e}")

    # Process with TotalSegmentator
    if config.use_totalsegmentator and TOTALSEGMENTATOR_AVAILABLE:
        try:
            seg_volume, elapsed = segment_with_totalsegmentator(image_path, output_dirs['ts'])

            if seg_volume is not None:
                # Save
                save_segmentation(seg_volume, original_affine, patient_id, 'ts', output_dirs)

                results['segmentations']['ts'] = seg_volume
                results['times']['ts'] = elapsed

            cleanup_memory()

        except Exception as e:
            print(f" TotalSegmentator failed: {e}")

    # Process with WholeBodyCT
    if config.use_wholebody_ct and wbct_components is not None:
        try:
            seg_volume, elapsed = segment_with_wholebody_ct(image_path, wbct_components)

            if seg_volume is not None:
                # Resize to original if needed
                if seg_volume.shape != original_shape:
                    seg_volume = resize_segmentation(seg_volume, original_shape)

                # Save
                save_segmentation(seg_volume, original_affine, patient_id, 'wbct', output_dirs)

                results['segmentations']['wbct'] = seg_volume
                results['times']['wbct'] = elapsed

            cleanup_memory()

        except Exception as e:
            print(f" WholeBodyCT failed: {e}")

    # Create visualizations
    if results['segmentations']:
        vis_path = output_dirs['visualizations'] / f"{patient_id}_comparison.png"
        visualize_segmentation(image_path, results['segmentations'], save_path=str(vis_path))

        grid_path = output_dirs['visualizations'] / f"{patient_id}_grid.png"
        create_comparison_grid(image_path, results['segmentations'], save_path=str(grid_path))

    # Summary
    print("\n Processing Summary:")
    for model_name, elapsed in results['times'].items():
        print(f"   {model_name.upper()}: {elapsed:.2f}s")

    return results

# Override the original function
process_patient = process_patient_with_wbct

print(" process_patient function updated with WholeBodyCT support")

# ============================================================================
# Ready Message
# ============================================================================

print("\n" + "="*70)
print(" WholeBodyCT Integration Complete!")
print("="*70)
print("\n To use WholeBodyCT, set these paths:")
print("   config.wbct_model_path = '/path/to/your/model.pth'")
print("   config.wbct_config_path = '/path/to/your/config.json'")
print("\nThen re-run this cell to load the model.")
print("\n You can now process images with all 3 models!")
print("="*70)
````

*   **Official Repository (GitHub):** The primary repository for MedSAM can be found at `https://github.com/bowang-lab/MedSAM`.

### Model Usage and Suitability for Whole-Body CT
MedSAM aims to provide zero-shot segmentation capabilities for various medical imaging modalities, including CT. While it can perform segmentation on CT scans, its strength lies in interactive or prompt-based segmentation, where a user provides a prompt (like a point or bounding box) to segment a specific region. For *fully automated* whole-body CT segmentation of *multiple organs without prompts*, direct application might require additional wrapper logic or fine-tuning compared to models explicitly trained for multi-organ segmentation (like TotalSegmentator or MONAI's `WholeBody_ct_segmentation` bundle).

### NIfTI Compatibility

Output segmentations can usually be saved back into NIfTI format using libraries like `nibabel`.

### MONAI Compatibility
MedSAM's architecture is based on standard deep learning components (e.g., Vision Transformers), making it compatible with PyTorch. ### Organ Mapping and Comprehensive Organ Sets
MedSAM is a *segment anything* model, meaning it doesn't have pre-defined classes for specific organs like BTCV. Instead, it segments *whatever* is prompted. To achieve multi-organ segmentation similar to BTCV, one would typically need:

1.  **Automated Prompt Generation:** A mechanism to generate prompts (e.g., bounding boxes or central points) for each target organ within the CT scan.
2.  **Iterative Segmentation:** Run MedSAM inference for each prompt.
3.  **Labeling:** Assign a specific label (e.g., BTCV organ ID) to each segmented region.

Therefore, MedSAM doesn't provide an inherent organ mapping. It requires external logic to identify and label organs, which adds complexity for fully automated multi-organ segmentation compared to pre-trained multi-organ models.

**Reasoning**:
With MedSAM2 dependencies installed and the model weights downloaded, the next step is to integrate the MedSAM2 model loading logic into the environment. This involves importing necessary components from the MedSAM2 library, defining a function to load the model and its weights, and then calling it to instantiate the model.



````python
import torch
import sys
from pathlib import Path

# Ensure MedSAM2 path is in sys.path (already done in previous cell, but good practice)
medsam2_path = '/content/MedSAM2'
if medsam2_path not in sys.path:
    sys.path.append(medsam2_path)

# Import MedSAM2 components
try:
    from sam_vit import SamPredictor, sam_model_registry
    print(" Successfully imported SamPredictor and sam_model_registry from MedSAM2.")
except ImportError as e:
    print(f" Failed to import MedSAM2 components: {e}")
    print("   Please ensure MedSAM2 is correctly installed and its path is in sys.path.")

def load_medsam2_model(checkpoint_path: str, model_type: str = "vit_b", device: str = "cuda"):
    """Load the MedSAM2 model with specified checkpoint and device."""
    print(f"\n Loading MedSAM2 model of type '{model_type}' from {checkpoint_path}...")
    if not Path(checkpoint_path).exists():
        print(f" Model checkpoint not found at {checkpoint_path}")
        return None

    try:
        sam_model = sam_model_registry[model_type](checkpoint=checkpoint_path)
        sam_model.to(device=device)
        sam_model.eval()
        print(" MedSAM2 model loaded successfully.")
        return SamPredictor(sam_model)
    except Exception as e:
        print(f" Error loading MedSAM2 model: {e}")
        return None

# Define the path to the downloaded MedSAM2 checkpoint
medsam2_checkpoint_path = "/content/MedSAM2/checkpoints/sam2.1_hiera_tiny.pt"

# Check if GPU is available
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

# Load the MedSAM2 model
medsam2_predictor = load_medsam2_model(medsam2_checkpoint_path, model_type="vit_b", device=device)

if medsam2_predictor:
    print("MedSAM2 model is ready for inference.")
else:
    print("MedSAM2 model could not be loaded. Please check the path and dependencies.")
````

````python
from pathlib import Path

dirs_to_check = [
    "/content/extracted_data/s0011/segmentations/swin_unetr/organs",
    "/content/extracted_data/s0011/segmentations/totalsegmentator/organs"
]

target_organs = ['kidney', 'stomach', 'liver']

for directory in dirs_to_check:
    print(f"\n--- Checking directory: {directory} ---")
    path = Path(directory)

    if not path.exists() or not path.is_dir():
        print(f" Directory not found or is not a directory: {directory}")
        continue

    found_files_in_dir = False
    for organ_keyword in target_organs:
        # Using glob to find files containing the organ keyword
        for file_path in path.glob(f"*{organ_keyword}*.nii.gz"):
            print(f" Found: {file_path.name}")
            found_files_in_dir = True

    if not found_files_in_dir:
        print(f"No files for kidney, stomach, or liver found in {directory}")
````

````python
import shutil
import os
from pathlib import Path

source_folder_to_zip = "/content/s0011/segmentations"
output_zip_name = "s0011_segmentations_archive"
output_zip_path = Path("/content") / f"{output_zip_name}.zip"

# Ensure the source directory exists
if not Path(source_folder_to_zip).is_dir():
    print(f" Error: The directory '{source_folder_to_zip}' does not exist.")
else:
    print(f"Compressing '{source_folder_to_zip}' into '{output_zip_path}'...")

    # Use shutil.make_archive to create a zip file
    # base_name: The name of the archive file to create, including the path
    # format: The archive format (e.g., 'zip', 'tar', 'gztar')
    # root_dir: The directory from which to start archiving. All paths in the archive will be relative to this directory.
    # base_dir: The directory that will be archived. All files and directories under base_dir will be added to the archive.
    shutil.make_archive(
        base_name=str(Path("/content") / output_zip_name), # This will create /content/s0011_segmentations_archive.zip
        format='zip',
        root_dir=Path(source_folder_to_zip).parent, # Parent directory of segmentations
        base_dir=Path(source_folder_to_zip).name # Just the 'segmentations' folder
    )

    if output_zip_path.exists():
        print(f" Successfully created zip file: {output_zip_path}")
        print("You can now download this file from the Colab file browser (folder icon on the left).")
    else:
        print(f" Failed to create zip file: {output_zip_path}")
````
